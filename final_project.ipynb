```python
# COMPLETE: train models, predict per-product future qty & sales, produce ranking
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

# 1) Load data
df = pd.read_csv("final_project.csv")

# 2) Basic cleaning and typing
df.columns = df.columns.str.strip().str.replace(" ", "_").str.lower()
# Ensure date
if "date" in df.columns:
    df["date"] = pd.to_datetime(df["date"], errors="coerce")
# numeric cols
for c in ["price", "quantity", "discount", "tax"]:
    if c in df.columns:
        df[c] = pd.to_numeric(df[c], errors="coerce")
# clean text
for c in ["product", "category", "city", "region", "payment_method"]:
    if c in df.columns:
        df[c] = df[c].astype(str).str.title().str.strip()

# 3) Create total_sales if not present or to ensure correct calculation
# Use formula price*quantity + tax - discount
if "total_sales" not in df.columns:
    df["total_sales"] = df["price"] * df["quantity"] + df.get("tax", 0) - df.get("discount", 0)
else:
    # ensure numeric
    df["total_sales"] = pd.to_numeric(df["total_sales"], errors="coerce")

# 4) Drop rows where essential values are missing
df = df.dropna(subset=["price", "quantity", "total_sales"])

# 5) Feature selection for models (numeric only for simplicity)
FEATURES = ["price", "discount", "tax"]   # features we will use
# If any feature missing, create it with zeros
for f in FEATURES:
    if f not in df.columns:
        df[f] = 0
# Keep only relevant cols
data = df.copy()

# 6) Train model to predict quantity (target = quantity)
X_qty = data[FEATURES]
y_qty = data["quantity"]

X_train_q, X_test_q, y_train_q, y_test_q = train_test_split(X_qty, y_qty, test_size=0.2, random_state=42)

rf_qty = RandomForestRegressor(n_estimators=200, random_state=42)
rf_qty.fit(X_train_q, y_train_q)
y_pred_q = rf_qty.predict(X_test_q)

print("=== Quantity model evaluation ===")
print("R2 (qty):", r2_score(y_test_q, y_pred_q))
print("MAE (qty):", mean_absolute_error(y_test_q, y_pred_q))
print("RMSE (qty):", np.sqrt(mean_squared_error(y_test_q, y_pred_q)))
print()

# 7) Train model to predict total_sales (target = total_sales)
X_sales = data[FEATURES]
y_sales = data["total_sales"]

X_train_s, X_test_s, y_train_s, y_test_s = train_test_split(X_sales, y_sales, test_size=0.2, random_state=42)

rf_sales = RandomForestRegressor(n_estimators=200, random_state=42)
rf_sales.fit(X_train_s, y_train_s)
y_pred_s = rf_sales.predict(X_test_s)

print("=== Sales model evaluation ===")
print("R2 (sales):", r2_score(y_test_s, y_pred_s))
print("MAE (sales):", mean_absolute_error(y_test_s, y_pred_s))
print("RMSE (sales):", np.sqrt(mean_squared_error(y_test_s, y_pred_s)))
print()

# 8) Create per-product "future" features for prediction
# We aggregate product-wise using median/mean of numeric features so model sees realistic inputs.
agg_funcs = {
    "price": "median",
    "discount": "median",
    "tax": "median"
}
product_features = data.groupby("product").agg(agg_funcs).reset_index()

# Optional: you may want to adjust price/discount/tax per expected future scenario.
# e.g., product_features['price'] *= 1.02  # assume 2% price increase next period

# 9) Predict quantity and sales per product using the trained models
X_future = product_features[["price", "discount", "tax"]].fillna(0)  # ensure no NaN
product_features["predicted_quantity"] = rf_qty.predict(X_future).round(2)
product_features["predicted_sales"] = rf_sales.predict(X_future).round(2)

# 10) Create ranking and combined score (optional)
# Rank by predicted_quantity and predicted_sales (you can pick either or combine)
product_features["qty_rank"] = product_features["predicted_quantity"].rank(ascending=False, method="min")
product_features["sales_rank"] = product_features["predicted_sales"].rank(ascending=False, method="min")

# Combined score (normalize then sum)  gives balanced ranking of both qty & value
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
norm_vals = scaler.fit_transform(product_features[["predicted_quantity", "predicted_sales"]])
product_features["combined_score"] = (norm_vals[:,0] + norm_vals[:,1])  # higher = better
product_features["combined_rank"] = product_features["combined_score"].rank(ascending=False, method="min")

# 11) Sort and show top products
product_features = product_features.sort_values("combined_rank")
cols_to_show = ["product", "price", "discount", "tax", "predicted_quantity", "predicted_sales", "combined_rank"]
print("=== Top products by combined rank ===")
print(product_features[cols_to_show].head(20).to_string(index=False))

# 12) Save results to CSV for review
product_features.to_csv("product_future_predictions.csv", index=False)
print("\nSaved product_future_predictions.csv (contains predictions + ranks).")
```
